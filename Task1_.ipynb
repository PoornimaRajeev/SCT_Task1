```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
# load dataset
df = pd.read_csv('/content/adult.csv')
df.head()
```

```python
print(df.columns.tolist())
```

```python
# Set figure size and style
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")

# Create histogram with smooth curve
sns.histplot(df['age'], bins=30, kde=True, color="#5DADE2", edgecolor="black", linewidth=1.2)

# Titles and labels
plt.title("📊 Age Distribution of Individuals", fontsize=16, weight='bold')
plt.xlabel("Age", fontsize=12)
plt.ylabel("Number of People", fontsize=12)

# Add grid, adjust layout
plt.grid(True, linestyle='--', alpha=0.5)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()

# Show
plt.show()
```
